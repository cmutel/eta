## Exercise 2: Different method

Switch to "Main air pollutants and PM" and report the score.

```python
lca.switch_method(METHODS[1])
lca.redo_lcia()
# What is the score?
print(f"Method: {METHODS[1][1]}")
print(f"Score: {lca.score:,.0f} UBP")
```

**Key insight:** `switch_method()` swaps the characterisation factors without rebuilding the inventory matrix — the LCI (supply chain calculation) is reused. Only `redo_lcia()` needs to run again. This is much faster than constructing a new `LCA` object for every method.

## Exercise 3: Dominant phase

Which life-cycle phase dominates for Land Use?

```python
# Construction phase dominates for Land Use due to stone and gravel extraction
```

**Explanation:** The Construction phase dominates Land Use because extracting raw materials — natural stone, gravel, and sand — requires large areas of quarrying and mining. These are one-time, high-mass inputs (over 5,000 kg of rock and aggregate). Land use impact methods account for both the area occupied and the time of occupation, so the extraction of inert materials has a disproportionately large land use footprint compared to the ongoing maintenance activities.
